```python
import copy

with open('data/day11.csv') as input:
    data = [[x for x in line] for line in input.read().splitlines()]

min_x = 0
min_y = 0
max_x = len(data[0])
max_y = len(data)

def get_adjacent(data, x, y):
    occupied = 0
    for j in range(-1, 2, 1):
        x_new = x + j
        if min_x <= x_new < max_x:
            for i in range(-1, 2, 1):
                y_new = y + i
                if min_y <= y_new < max_y and not (i == 0 and j == 0):
                    occupied += data[y_new][x_new].count('#')
    return occupied

finished = False

while not finished:
    data_new = copy.deepcopy(data)
    for j in range(max_x):
        for i in range(max_y):
            occupied = get_adjacent(data, j, i)
            cur_value = data[i][j]
            if occupied == 0 and cur_value != '.':
                data_new[i][j] = '#'
            elif occupied >= 4 and cur_value != '.':
                data_new[i][j] = 'L'
    # print(data_new)
    if data == data_new:
        finished = True
    else:
        data = copy.deepcopy(data_new)

# print(data)

seats_taken = 0
for row in data:
    seats_taken += row.count('#')

print(seats_taken)
```
