```python
with open('data/day03.csv', 'r') as input:
    data = [x for x in input.read().splitlines()]
    print(data)
```

```python
step_size_right = 3
step_size_down = 1
height = len(data)
width = len(data[0])
treesfound = 0

for i in range(step_size_down, height, step_size_down):
    if data[i][(step_size_right * i) % width] == '#':
        treesfound += 1

print(treesfound)
```

```python
step_size_array = [[1, 1], [3, 1], [5, 1], [7, 1], [1, 2]]
height = len(data)
width = len(data[0])
treesfoundproduct = 1

for j in range(len(step_size_array)):
    treesfound = 0
    step_size_right, step_size_down = step_size_array[j]
    for i in range(step_size_down, height, step_size_down):
        if data[i][(step_size_right * (i // step_size_down)) % width] == '#':
            treesfound += 1
    treesfoundproduct *= treesfound

print(treesfoundproduct)
```
