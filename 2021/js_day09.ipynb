```python
import time
start = time.time()

with open('js_data/day09.csv','r') as data:
    rows = data.read().split('\n')
    dataset1 = [[int(num) for num in '9'+row+'9'] for row in rows]
    dataset2 = [[int(num) for num in '0'+row+'0'] for row in rows]

dataset1.insert(0, [9 for x in range(len(dataset1[0]))])
dataset1.append([9 for x in range(len(dataset1[0]))])

dataset2.insert(0, [0 for x in range(len(dataset2[0]))])
dataset2.append([0 for x in range(len(dataset2[0]))])

risk = 0
to_check = {}
checked = {}
size = {}
values = []

for i in range(len(dataset1)-1):
    for j in range(len(dataset1[i])-1):
        value   = dataset1[i][j]
        north   = dataset1[i-1][j]
        east    = dataset1[i][j+1]
        south   = dataset1[i+1][j]
        west    = dataset1[i][j-1]
        if value < north and value < east and value < south and value < west:
            to_check[(i, j)] = [[i, j]]
            checked[(i, j)] = []
            size[(i, j)] = 0
            risk += value + 1

print(risk)

def lowest_point_check_basin(list):
    i_risk = list[0][0]
    j_risk = list[0][1]
    size_basin = 1
    while len(list) > 0:
        checked[(i_risk, j_risk)].append([list[0][0], list[0][1]])
        size_basin += check_basin(list[0][0], list[0][1], i_risk, j_risk)
        #print(f"checked[({i_risk}, {j_risk})]: {checked[(i_risk, j_risk)]}")
    return size_basin

def check_basin(i, j, i_risk, j_risk):
    value   = dataset2[i][j]
    north   = dataset2[i-1][j]
    east    = dataset2[i][j+1]
    south   = dataset2[i+1][j]
    west    = dataset2[i][j-1]
    added_nodes = 0

    if value < north and not [i-1, j] in checked[(i_risk, j_risk)]:
        if not [i-1, j] in to_check[(i_risk, j_risk)]:
            if north <= 8:
                to_check[(i_risk, j_risk)].append([i-1, j])
                added_nodes += 1
    if value < east and not [i, j+1] in checked[(i_risk, j_risk)]:
        if not [i, j+1] in to_check[(i_risk, j_risk)]:
            if east <= 8:
                to_check[(i_risk, j_risk)].append([i, j+1])
                added_nodes += 1
    if value < south and not [i+1, j] in checked[(i_risk, j_risk)]:
        if not [i+1, j] in to_check[(i_risk, j_risk)]:
            if south <= 8:
                to_check[(i_risk, j_risk)].append([i+1, j])
                added_nodes += 1
    if value < west and not [i, j-1] in checked[(i_risk, j_risk)]:
        if not [i, j-1] in to_check[(i_risk, j_risk)]:
            if west <= 8:
                to_check[(i_risk, j_risk)].append([i, j-1])
                added_nodes += 1
    to_check[(i_risk, j_risk)].remove([i, j])
    return added_nodes

# print(to_check)
# print(risk)

# print(to_check[(1, 2)])
# print(size)
# lowest_point_check_basin(to_check[(1, 2)])
# print(f"to check: {to_check}")

for key in size:
    size[key] += lowest_point_check_basin(to_check[key])
    values.append(int(size[key]))
    #print(f"{key}: {size[key]}")
    #print('\n')

#print(size)
values.sort(reverse=True)
#print(values)
print(values[0], values[1], values[2], values[0] * values[1] * values[2])
print(f"runtime: {time.time() - start}")

# '6~22'
# '12~70'
# '48~70'


#TOO LOW 539448
```
