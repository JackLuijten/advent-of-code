```python
import time
import matplotlib.pyplot as plt

starttime = time.time()

with open('data/day13.csv','r') as f:
    data = f.read().split('\n\n')
    pointsList = [[int(j) for j in i.split(',')] for i in data[0].split('\n')]
    folds = data[1].split('\n')
    points = set()
    for point in pointsList:
        i, j = point
        points.add((i, j))

for fold in folds:
    loc = int(fold.split('=')[1])
    if 'fold along y' in fold:
        changes = [(i, j) for (i, j) in points if j > loc]
        print(fold, changes)
        for e in changes:
            i, j = e
            points.remove((i, j))
            points.add((i, 2 * loc - j))
    if 'fold along x' in fold:
        changes = [(i, j) for (i, j) in points if i > loc]
        print(fold, changes)
        for e in changes:
            i, j = e
            points.remove((i, j))
            points.add((2 * loc - i, j))
    # print(fold.split('=')[1])

x = [x[0] for x in points]
y = [y[1] for y in points]
plt.ylim(-1,35)
plt.scatter(x, y)

print(len(points))
print(points)
print(folds)
```
