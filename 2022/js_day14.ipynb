```python
import time

starttime = time.time()

with open('data/day14.csv') as f:
    data = [[[n.split(',') for n in i.split(', ')] for i in j.split(' -> ')] for j in f.read().split('\n')]
    print(data)

wall = set()
sand = set()
sand2 = set()
ymax = 0

for i in data:
    # print(i)
    for j in range(len(i) - 1):
        x1, y1 = int(i[j][0][0]), int(i[j][0][1])
        x2, y2 = int(i[j+1][0][0]), int(i[j+1][0][1])
        ymax = max(ymax, y1, y2)
        # print((x1, y1), (x2, y2))
        if x1 == x2:
            direction = 1 if y2 > y1 else -1
            for k in range(abs(y2 - y1) + 1):
                wall.add((x1, y1 + (k * direction)))
        if y1 == y2:
            direction = 1 if x2 > x1 else -1
            for k in range(abs(x2 - x1) + 1):
                wall.add((x1 + (k * direction), y1))

print(len(wall))
print(ymax)
deepestSand = 0
NumberOfSandUnits = 0

def dropSand1(sandUnit):
    sandMoved = True
    global deepestSand
    while sandMoved:
        nextSand = (sandUnit[0], sandUnit[1] + 1)
        nextSandLeft = (sandUnit[0] - 1, sandUnit[1] + 1)
        nextSandRight = (sandUnit[0] + 1, sandUnit[1] + 1)
        if sandUnit[1] >= ymax:
            deepestSand = sandUnit[1]
            # print(4)
            break
        elif not (nextSand in wall or nextSand in sand):
            # print(1)
            sandUnit = nextSand
        elif not (nextSandLeft in wall or nextSandLeft in sand):
            # print(2)
            sandUnit = nextSandLeft
        elif not (nextSandRight in wall or nextSandRight in sand):
            # print(3)
            sandUnit = nextSandRight

        else:
            # print(5)
            sand.add(sandUnit)
            sandMoved = False
            deepestSand = max(deepestSand, sandUnit[1])

def dropSand2(sandUnit):
    sandMoved = True
    while sandMoved:
        nextSand = (sandUnit[0], sandUnit[1] + 1)
        nextSandLeft = (sandUnit[0] - 1, sandUnit[1] + 1)
        nextSandRight = (sandUnit[0] + 1, sandUnit[1] + 1)
        if sandUnit[1] == ymax + 1:
            sand2.add(sandUnit)
            sandMoved = False
            # print(4, sandUnit)
        elif not (nextSand in wall or nextSand in sand2):
            # print(1, sandUnit)
            sandUnit = nextSand
        elif not (nextSandLeft in wall or nextSandLeft in sand2):
            # print(2, sandUnit)
            sandUnit = nextSandLeft
        elif not (nextSandRight in wall or nextSandRight in sand2):
            # print(3, sandUnit)
            sandUnit = nextSandRight
        else:
            # print(5, sandUnit)
            sand2.add(sandUnit)
            sandMoved = False

while deepestSand < ymax:
    sandUnit = (500, 0)
    dropSand1(sandUnit)

print(len(sand))

while not (500, 0) in sand2:
    sandUnit = (500, 0)
    dropSand2(sandUnit)

print(len(sand2))

print(f"runtime: {time.time() - starttime}")
```
