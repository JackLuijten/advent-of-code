```python
import re

with open('data/day16.csv') as f:
    data = f.read().split('\n')

vertexValue = {}
graph = {}
for row in data:
    match = re.search(r"Valve (\w+).*=(-?\d+).*valves? (-?.*)", row)
    source, rate, dest = match.groups()
    graph[source] = set(dest.split(', '))
    if int(rate) > 0 or source == 'AA':
        vertexValue[source] = int(rate)

def find_distance(graph, start):
    dist = {start: 0}
    queue = []
    queue.append(start)
    while queue:
        curr = queue.pop(0)
        for neighbour in graph[curr]:
            if neighbour not in dist:
                dist[neighbour] = dist[curr] + 1
                queue.append(neighbour)
    return dist

interestingGraph = {}
interestingVertices = [i for i in vertexValue.keys()]
for vertex in vertexValue:
    interestingGraph[vertex] = find_distance(graph,vertex)

newGraph = {}
for vertex in interestingVertices:
    newGraph[vertex] = set(i for i in interestingVertices if i != vertex)

values = [0]

def depthFirst(graph=newGraph, currentVertex='AA', visited=[], timespent=0, value=0):
    visited.append(currentVertex)
    if not currentVertex == 'AA':
        timeleft = 30 - timespent
        value += timeleft * vertexValue[currentVertex]
    distances = interestingGraph[currentVertex]
    for vertex in graph[currentVertex]:
        if timespent + distances[vertex] >= 30 or len(visited) == len(interestingVertices):
            if value > max(values):
                values.append(value)
            continue
        elif not vertex in visited:
            depthFirst(graph, vertex, visited.copy(), timespent + distances[vertex] + 1, value)

visitedList2 = []
values2 = [0]

def depthSecond(graph=newGraph, currentVertex1='AA', currentVertex2='AA', visited=[], timespent1=0, timespent2=0, value=0, person=1):
    if person == 1:
        visited.append(currentVertex2)
        if not currentVertex1 == 'AA':
            timeleft = 26 - timespent2
            value += timeleft * vertexValue[currentVertex2]
        distances = interestingGraph[currentVertex1]
        for vertex in graph[currentVertex1]:
            if timespent1 + distances[vertex] >= 26 or len(visited) == len(interestingVertices):
                if value > max(values2):
                    values2.append(value)
                continue
            elif not vertex in visited:
                depthSecond(graph, vertex, currentVertex2, visited.copy(), timespent1 + distances[vertex] + 1, timespent2, value, 2)
    elif person == 2:
        visited.append(currentVertex1)
        if not currentVertex1 == 'AA':
            timeleft = 26 - timespent1
            value += timeleft * vertexValue[currentVertex1]
        distances = interestingGraph[currentVertex2]
        for vertex in graph[currentVertex2]:
            if timespent2 + distances[vertex] >= 26 or len(visited) == len(interestingVertices):
                if value > max(values2):
                    values2.append(value)
                continue
            elif not vertex in visited:
                depthSecond(graph, currentVertex1, vertex, visited.copy(), timespent1, timespent2 + distances[vertex] + 1, value, 1)

depthFirst()
print(max(values))

print('\n')

# depthSecond()
# print(max(values2))
```
